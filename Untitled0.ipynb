```python
''' Importe de librerias '''

import pandas as pd
import matplotlib.pyplot as plt

''' carga de datos '''

store1 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv")
store2 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv")
store3 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv")
store4 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv")
```

```python
''' ingreso total por tienda '''

ingreso1 = store1["Precio"].sum()
ingreso2 = store2["Precio"].sum()
ingreso3 = store3["Precio"].sum()
ingreso4 = store4["Precio"].sum()

print("Ingreso tienda 1:", ingreso1)
print("Ingreso tienda 2:", ingreso2)
print("Ingreso tienda 3:", ingreso3)
print("Ingreso tienda 4:", ingreso4)
```

```python
''' ventas por categoria '''

ventas_categoria1 = store1["Categoría del Producto"].value_counts()
ventas_categoria2 = store2["Categoría del Producto"].value_counts()
ventas_categoria3 = store3["Categoría del Producto"].value_counts()
ventas_categoria4 = store4["Categoría del Producto"].value_counts()

print(ventas_categoria1)
```

```python
''' valoracion media por tienda '''

valoracion1 = store1["Calificación"].mean()
valoracion2 = store2["Calificación"].mean()
valoracion3 = store3["Calificación"].mean()
valoracion4 = store4["Calificación"].mean()

print("Valoración tienda 1:", valoracion1)
print("Valoración tienda 2:", valoracion2)
print("Valoración tienda 3:", valoracion3)
print("Valoración tienda 4:", valoracion4)
```

```python
''' productos mas y menos vendidos '''

productos_tienda1 = store1["Producto"].value_counts()

print("Más vendido:")
print(productos_tienda1.head(1))

print("Menos vendido:")
print(productos_tienda1.tail(1))
```

```python
''' valor del envio promedio por tiendas '''

envio1 = store1["Costo de envío"].mean()
envio2 = store2["Costo de envío"].mean()
envio3 = store3["Costo de envío"].mean()
envio4 = store4["Costo de envío"].mean()

print(envio1, envio2, envio3, envio4)
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso1, ingreso2, ingreso3, ingreso4]

plt.bar(tiendas, ingresos)
plt.title("Ingresos por tienda")
plt.ylabel("Ingresos")
plt.show()
```

```python
ventas_categoria1.head().plot(kind="bar")

plt.title("Categorías más vendidas - Tienda 1")
plt.ylabel("Cantidad de ventas")
plt.show()
```

```python
valoraciones = [valoracion1, valoracion2, valoracion3, valoracion4]

plt.scatter(tiendas, valoraciones)

plt.title("Valoración promedio por tienda")
plt.ylabel("Calificación")
plt.show()
```

# Informe

El objetivo de este análisis es evaluar el desempeño de las cuatro tiendas para determinar cuál de ellas debería vender el Sr. Juan con el fin de iniciar un nuevo emprendimiento. Para tomar esta decisión, se analizaron diferentes indicadores de rendimiento de cada tienda, incluyendo los ingresos totales, las categorías de productos más y menos vendidas, las calificaciones promedio de los clientes, los productos más y menos vendidos y el costo promedio de envío.

# Ingresos totales

El análisis de los ingresos totales permitió identificar qué tienda genera mayor volumen de ventas. Las tiendas con mayores ingresos demuestran un mejor desempeño comercial y mayor capacidad para atraer clientes.

# Categoría de productos

El análisis de las categorías de productos permitió identificar cuáles son las más populares en cada tienda. Las categorías más vendidas reflejan las preferencias de los clientes y pueden indicar oportunidades de crecimiento.

# Calificación promedio de los clientes

La satisfacción del cliente es un factor clave para evaluar el rendimiento de una tienda. Para ello se analizó la calificación promedio de los clientes en cada tienda.

# Productos más y menos vendidos

También se analizaron los productos con mayor y menor número de ventas en cada tienda. Este análisis permite identificar los productos más populares entre los clientes, así como aquellos que presentan menor demanda.

# Costo promedio de envío

Otro factor analizado fue el costo promedio de envío en cada tienda. Dado que el cliente es quien asume este costo, el valor del envío puede influir en la decisión de compra.

# Conclusión y recomendación

Con base en los análisis realizados y en las visualizaciones generadas, se recomienda que el Sr. Juan considere vender la tienda que presenta el menor rendimiento general en comparación con las demás.

Esta recomendación se basa en factores como menores ingresos totales, menor volumen de ventas en las principales categorías de productos, calificaciones promedio más bajas por parte de los clientes y menor desempeño en los productos más vendidos.
